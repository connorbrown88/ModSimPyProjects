```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import math
```

```python
def round_down(n, decimals=0):
    multiplier = 10 ** decimals
    return math.floor(n * multiplier) / multiplier

def round_up(n, decimals=0):
    multiplier = 10 ** decimals
    return math.ceil(n * multiplier) / multiplier
```

```python
series = pd.read_csv('test_data.csv', header = None)
```

```python
series
```

```python
for points in series[0]:
    
    plt.figure()
    
    ax = plt.subplot(projection='polar')
    
    #Remove Black Outline
    plt.box(on=None)
    
    #Generate Light Blue Bar
    ax.barh(2, math.radians(180), alpha = 0.15, color = 'blue')
    
    #Generate Points Bar
    rad = ((points - round_down(points, -3)) / 1000) * 180
    
    ax.barh(2, math.radians(rad), alpha = 0.8, color = 'blue')
    ax.set_theta_zero_location('W')
    ax.set_theta_direction(-1)
    
    #labels = will pass in arguments
    lower_bound = int(round_down(points, -3))
    upper_bound = int(round_up(points, -3))
    if points - lower_bound == 0:
        upper_bound = upper_bound + 1000

    ax.set_thetagrids([0, 180], labels=[lower_bound, upper_bound], fontweight='bold', color = 'blue', verticalalignment = 'bottom')
    ax.set_rgrids([], labels=[])
    ax.grid(False)
    
    #Select Arial as chosen font
    csfont = {'fontname':'Arial'}
    
    #Create Title
    ax.set_title('Total Points Progress', **csfont, color = 'blue', fontsize=18, fontweight='bold')
    
    #Create Points Text
    ax.text(0.5, 0.6, str(points), ha='center', va='center', **csfont, color = 'blue', fontsize=18, fontweight='bold', transform=ax.transAxes)
    
    
    plt.show()
```
