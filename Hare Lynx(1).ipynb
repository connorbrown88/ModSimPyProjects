```python
#Takes in the series, initial year, and current year, and computes the average value of the min peaks
# ** Defines a minpeak as being less than the previous 2 AND the next 2

def minAverage(series, initYear, curYear):        
    
    minPeaks = []
    
    for i in range((initYear + 2), (1933)):
        if (series[i] < series[i-1]):
            if(series[i] < series[i-2]):
                if(series[i] < series[i+1]):
                    if(series[i] < series[i+2]):
                        minPeaks.append(series[i])

    return np.mean(minPeaks)
```

```python
#Takes in the series, initial year, and current year, and computes the average value of the max peaks
# ** Defines a maxpeak as being greater than the previous 2 AND the next 2

def maxAverage(series, initYear, curYear):
    
    maxPeaks = []
    
    for i in range((initYear + 2), (1933)):
        if (series[i] > series[i-1]):
            if(series[i] > series[i-2]):
                if(series[i] > series[i+1]):
                    if(series[i] > series[i+2]):
                        maxPeaks.append(series[i])
    
    return np.mean(maxPeaks)
```

```python
def choose_Pos_Slope(series, curRatio, initYear, curYear, prevRatio):
    if (curRatio < minAverage(series, initYear, curYear)):
        return True
    if (curRatio > maxAverage(series, initYear, curYear)):
        return False
    if (curRatio > minAverage(series, initYear, curYear) and curRatio > prevRatio):             
        return True
    else:
        return False
```

```python
#Takes in the series, current ratio of the year, the initial year, current year, checks to see if the positive slope
#or negative slope should be chosen, then returns either the mean positive slope, or mean negative slope. 

def calc_mean_Slope(series, curRatio, initYear, curYear, prevRatio):
    Slopes = []
   
    
    if(choose_Pos_Slope(series, curRatio, initYear, curYear, prevRatio)):
        if(curRatio <= series.min()):                                           #(for when curRatio is below the lowest min)
            return 1.3
        
        for i in range(initYear, curYear):
            if(curRatio >= series[i] and curRatio <= series[i+1]):
                Slopes.append(series[i+1] - series[i])                          #Try to edit so it calculates the slope, and also considers the year intervals, right now it's assumed the year interval is 1.
        return np.mean(Slopes)
      
    
    else:
        if (curRatio >= series.max()):                                          #(for when curRatio is over the highest max)
            return -1.3
        
        for i in range(initYear, curYear):
            if(curRatio <= series[i] and  curRatio >= series[i+1]):
                Slopes.append(series[i+1] - series[i])
        return np.mean(Slopes)
```

```python

def run_simulation(series, curRatio, initYear, curYear, numTimes):
    prevRatio = 0 
    for i in range(numTimes):
        slope = calc_mean_Slope(series, curRatio, initYear, curYear, prevRatio)       #Returns the mean slope for that current ratio
        prevRatio = curRatio                                                          #Takes the curRatio and declares it prevRatio 
        curRatio += slope                                                             #Adds the slope(change in growth) to the previous curRatio
        series[curYear + 1] = curRatio                                                #Takes the new current ratio and passes it into the series
        curYear += 1                                                                  #Adds 1 to current Year
    
    return series
```
